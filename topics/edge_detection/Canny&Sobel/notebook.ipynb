# Sobel Edge Detector

![Sobel](Sobel1.png) ![Sobel2](Sobel2.png)


The operator uses two 3Ã—3 kernels which are convolved with the original image to calculate approximations of the **derivatives**

 
[1] Sobel, Irwin. "History and definition of the sobel operator." Retrieved from the World Wide Web 1505 (2014).





## Horizontal and Vertical Detector

One for horizontal changes, and one for vertical

![Sobel3](Sobel3.png)

## Magnitude

![Sobel4](Sobel4.png)

![Sobel5](Sobel5.png)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image
image = cv2.imread('test.jpeg', cv2.IMREAD_GRAYSCALE)
image = cv2.resize(image, (224, 224))

# Kernel Size
ksize = 3

# Apply Sobel operator in both horizontal and vertical directions
sobelx = cv2.Sobel(image, cv2.CV_64F, dx=1, dy=0, ksize=ksize)
sobely = cv2.Sobel(image, cv2.CV_64F, dx=0, dy=1, ksize=ksize)

# Calculate the magnitude
sobelMagnitude = np.sqrt(sobelx**2 + sobely**2)

# Normalize to 0-255
sobelMagnitude = np.uint8(255 * sobelMagnitude / np.max(sobelMagnitude))

# Display images using matplotlib
plt.figure(figsize=(10, 3))
plt.subplot(1, 4, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(sobelx, cmap='gray')
plt.title('Gradient X')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(sobely, cmap='gray')
plt.title('Gradient Y')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(sobelMagnitude, cmap='gray')
plt.title('Magnitude')
plt.axis('off')

plt.suptitle(f'kernel size = {ksize}', fontweight='bold')
```

# Canny Edge Detector
![Canny1](Canny1.png)

[1]Canny J. A computational approach to edge detection[J]. IEEE Transactions on pattern analysis and machine intelligence, 1986 (6): 679-698.

Canny detector is a multi-stage algorithm

### The process of Canny edge detection algorithm:
1. Apply Gaussian filter to smooth the image. -> Noise Reduction
2. Finding Intensity Gradient of the image -> Magnitude and Direction
3. Non-maximum Suppression -> Make Edge Thinner
4. Double Thresholding -> Potential edges are determined by two thresholds: a high and a low threshold

## Finding Intensity Gradient of the image

![Canny2](Canny2.png)

![Canny3](Canny3.png)



## Non-maximum Suppression

After getting gradient magnitude and direction, a full scan of image is done to remove any unwanted pixels which may not constitute the edge. For this, at every pixel, pixel is checked if it is a local maximum in its neighborhood in the direction of gradient.

![Canny4](Canny4.png)

## Double Threshold

This stage decides which are all edges are really edges and which are not. For this, we need two threshold values, minVal and maxVal. Any edges with intensity gradient more than maxVal are sure to be edges and those below minVal are sure to be non-edges, so discarded. Those who lie between these two thresholds are classified edges or non-edges based on their connectivity. If they are connected to "sure-edge" pixels, they are considered to be part of edges. Otherwise, they are also discarded. 

See the image below:
![Canny5](Canny5.png)

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def apply_canny_edge_detection(image_path, low_threshold, high_threshold):
    # Read the image
    img = cv2.imread(image_path, 0)  # 0 to read image in grayscale
    
    # Apply GaussianBlur to reduce noise
    blurred_img = cv2.GaussianBlur(img, (5, 5), 0)
    
    # Apply Canny edge detection
    edges = cv2.Canny(blurred_img, low_threshold, high_threshold)
    
    # Display the original and edged images
    plt.figure(figsize=(10, 5))
    plt.subplot(121), plt.imshow(img, cmap='gray')
    plt.title('Original Image'), plt.xticks([]), plt.yticks([])
    plt.subplot(122), plt.imshow(edges, cmap='gray')
    plt.title(f'Canny Edge Image, low_thres={low_threshold}, high_thres={high_threshold}'), plt.xticks([]), plt.yticks([])
    plt.show()


# Example
low_threshold = 50
high_threshold = 150
apply_canny_edge_detection('test.jpeg', low_threshold, high_threshold)
```

```python
for low_threshold in [25,50,75]:
    for high_threshold in [125,150,175]:
        apply_canny_edge_detection('test.jpeg', low_threshold, high_threshold)
```
