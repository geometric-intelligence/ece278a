# Blurred QR Code Decode Helper

## Shang-Hsun Yang

### Date: June 3, 2024

## Introduction

### **Goal:** Extract the link behind a blurred QR code

### **Situations:**

* Low-Resolution Printing
* Improper Printer Settings
* Unsuitable Printing Materials
* Environmental Factors

### **Importance:** QR codes are everywhere nowadays!

## Method

### **Tool:** Python + PIL (Python Imaging Library) package

### **Steps:**

1. Generate the QR code: `qrcode.QRCode()`
2. Blur the QR code: `ImageFilter.BoxBlur`
3. Sharpen the QR code: `PIL.ImageFilter.UnsharpMask` and `PIL.ImageEnhance.Contrast`
4. Decode the result: `BarCodeReader().decode`
